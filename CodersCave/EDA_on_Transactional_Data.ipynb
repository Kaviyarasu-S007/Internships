```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules

# Load the data from Excel into a Pandas DataFrame
file_path =  "Hackathon_Ideal_Data.csv"# Replace with the actual path to your Excel file
df = pd.read_csv(file_path)

# Data Cleaning
cleaned_data = df.dropna()  # Drop rows with missing values

# Exploratory Data Analysis (EDA)

# Distribution of Quantity - Pie Chart
quantity_distribution = cleaned_data['MONTH'].value_counts()
plt.pie(quantity_distribution, labels=quantity_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Quantity (Pie Chart)')
```

```python
# Distribution of Value - Stacked Bar Chart
sns.histplot(cleaned_data, x='VALUE', hue='MONTH', bins=20, multiple="stack", kde=True)
plt.title('Distribution of Value (Stacked Bar Chart by Month)')
```

```python

# Sales Distribution Across Months - Violin Plot
sns.violinplot(x=cleaned_data['MONTH'], y=cleaned_data['VALUE'], inner='quartile')
plt.title('Sales Distribution Across Months (Violin Plot)')
```

```python
# Store-wise Performance - Grouped Bar Chart
sns.barplot(x='STORECODE', y='VALUE', hue='MONTH', data=cleaned_data, ci=None)
plt.title('Store-wise Performance (Grouped Bar Chart)')
plt.xlabel('Store Code')
plt.ylabel('Total Sales')
```

```python
brand_distribution = cleaned_data['GRP'][:1000].value_counts()
plt.pie(brand_distribution, labels=brand_distribution.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.gca().add_artist(plt.Circle((0, 0), 0.3, fc='white'))
plt.title('Brand Analysis (Donut Plot)')
```

```python
# Sales Distribution Across Categories
plt.figure(figsize=(14, 8))
sns.histplot(data=df[:2000], x='VALUE', hue='GRP', multiple='stack', kde=True)
plt.title('Sales Distribution Across Product Groups')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(14, 8))
sns.histplot(data=df[:2000], x='VALUE', hue='SGRP', multiple='stack', kde=True)
plt.title('Sales Distribution Across Product Sub-Groups')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Contingency Table - Relationship between Brand (BRD) and Product Group (GRP)
contingency_table = pd.crosstab(df['BRD'][:500], df['GRP'][:500])
print("\nContingency Table - Brand vs Product Group:")
print(contingency_table)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['QTY'][:1000], df['VALUE'][:1000], df['VALUE'][:1000], c='skyblue', s=20)
ax.set_xlabel('Quantity')
ax.set_ylabel('Value')
ax.set_zlabel('Month')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Contingency Table Heatmap with Annotations')
plt.show()
```

```python
correlation_matrix = df[['QTY', 'VALUE', 'MONTH']][:1000].corr()
sns.clustermap(correlation_matrix, annot=True, cmap='coolwarm', figsize=(10, 8))
plt.title('Cluster Map for Correlation')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df[['QTY', 'VALUE', 'MONTH', 'BRD']][:1000], hue='BRD', diag_kind='kde')
plt.suptitle('Pairplot with Categorical Hue', y=1.02)
plt.show()
```
